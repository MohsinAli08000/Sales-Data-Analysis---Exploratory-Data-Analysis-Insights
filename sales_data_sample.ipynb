```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding="ISO-8859-1")
```

```python
df.fillna(0, inplace=True)
```

```python
df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
df['ORDERDATE']=pd.to_datetime(df['ORDERDATE'])
df['ORDERDATE'].fillna(method='ffill', inplace=True)
df['ORDERDATE'].fillna(method='bfill', inplace=True)
```

```python
df['Month'] = df['ORDERDATE'].dt.to_period('M')
```

```python
monthly_sale=df.groupby('Month')['SALES'].sum()
```

```python
# Sales overtime
plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_sale.index.astype(str),y=monthly_sale.values,markers='+')
plt.xticks(rotation=40)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
#Sales by productline
plt.figure(figsize=(10,6))
sns.barplot(x=df['PRODUCTLINE'],y=df['SALES'],estimator=sum,ci=None)
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.title('Sales By Products ')
plt.show()
```

```python
#sales by country
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='COUNTRY', y='SALES', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Sales')
plt.title('Sales by Country')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='STATUS', palette='Set2', order=df['STATUS'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel("Order Status")
plt.ylabel("Count")
plt.title("Order Status Distribution")
plt.show()
```

```python
#Relationship Between Quantity ordered & SAles
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='QUANTITYORDERED',y='SALES',alpha=0.9)
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Quantity order vs. SAles')
plt.grid()
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,7))
sns.heatmap(df[['SALES', 'QUANTITYORDERED', 'PRICEEACH']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between SAles , Quantity & Price')
plt.show()
```

```python
#Top 10 Customer SAles
top_customers = df.groupby('CUSTOMERNAME')['SALES'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_customers.index,y=top_customers.values,palette='magma')
plt.xticks(rotation=45)
plt.xlabel('Customer Name')
plt.ylabel('Sales')
plt.title('Top 10 Customers by Total Sales')
plt.show()
```
